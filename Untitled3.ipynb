```python
Y=VeriSetim.iloc[:,9]
X=VeriSetim.drop(columns=['Potability'],axis=1)
X=VeriSetim[['Solids','Chloramines']]
```

```python
from sklearn.model_selection import train_test_split
XTrain,XTest,YTrain,YTest=train_test_split(X,Y,test_size=0.25,random_state=54)
```

```python
XTest
```

```python
#Logistic Regression

from sklearn.linear_model import LogisticRegression
classLG=LogisticRegression(random_state=0)
classLG.fit(XTrain,YTrain)
```

```python
PredictionLG=classLG.predict(XTest)
```

```python
from sklearn.metrics import confusion_matrix
cmLG=confusion_matrix(YTest,PredictionLG)
print(cmLG)
```

```python
import sklearn.metrics as mt
print(mt.accuracy_score(YTest,PredictionLG))
```

```python
#KNN(K en yakın komşu)
from sklearn.neighbors import KNeighborsClassifier
classKNN=KNeighborsClassifier(n_neighbors=10,metric='minkowski',p=2)
classKNN.fit(XTrain,YTrain)
```

```python
predKNN=classKNN.predict(XTest)
```

```python
from sklearn.metrics import confusion_matrix
cmKNN=confusion_matrix(YTest,predKNN)
print(cmKNN)
```

```python
import sklearn.metrics as mt
print(mt.accuracy_score(YTest,predKNN))
```

```python
#Karar Ağacı
from sklearn.tree import DecisionTreeClassifier
classKA=DecisionTreeClassifier(criterion='entropy',random_state=0)
classKA.fit(XTrain,YTrain)
```

```python
PredictionKA=classKA.predict(XTest)
from sklearn.metrics import confusion_matrix
cmKA=confusion_matrix(YTest,PredictionKA)
print(cmKA)
```

import sklearn.metrics as mt

```python
from sklearn import svm
classSVM=svm.SVC(kernel='linear')
classSVM.fit(XTrain,YTrain)
```

```python
import sklearn.metrics as mt
print(mt.accuracy_score(YTest,PredictionSVM))
```

```python
#Rasgele Orman
from sklearn.ensemble import RandomForestClassifier
classRF=RandomForestClassifier(n_estimators=1000, random_state=0)
classRF.fit(XTrain,YTrain)
PredictionRF=classRF.predict(XTest)
from sklearn.metrics import confusion_matrix
cmRF=confusion_matrix(YTest,PredictionRF)
print(cmRF)
import sklearn.metrics as mt
print(mt.accuracy_score(YTest,PredictionRF))
```
